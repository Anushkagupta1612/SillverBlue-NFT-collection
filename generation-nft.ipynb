```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%


face = ["yellow-face", "brown-face", "light-brown-face", "fair-face", "colored-face", "dull-brown-face", "green-face"]
face_weights = [35, 20, 15, 11, 9, 8, 2]

eyes = ["black-eyes", "white-eyes", "big-black-eyes", "green-eyes", "yellow-eyes", "patch-eyes", "goggle-eyes"]
eyes_weights = [35, 20, 15, 11, 9, 8, 2]

hair = ["normal-hair", "hair-band", "purple-hair", "multi-color", "red-white", "cap", "helmet"]
hair_weights = [35, 20, 15, 11, 9, 8, 2]

acc = ["earring", "cigar", "yellow-cigar", "mask", "yellow-dot", "smoking", "headphones"]
acc_weights = [35, 20, 15, 11, 9, 8, 2]
```

```python
#Classify traits

face_files ={
              "yellow-face" : "face_1", 
              "brown-face" : "face_2", 
              "light-brown-face" : "face_3" , 
              "fair-face" : "face_4" , 
              "colored-face" : "face_5" , 
              "dull-brown-face" : "face_6" , 
              "green-face" : "face_7"
            }

eyes_files = {"black-eyes" : "eye_0", 
        "white-eyes" : "eye_1", 
        "big-black-eyes" : "eye_2", 
        "green-eyes" : "eye_4", 
        "yellow-eyes" : "eye_5", 
        "patch-eyes" : "eye_6", 
        "goggle-eyes" : "eye_7"}

hair_files = {"normal-hair" : "hair_0", 
        "hair-band" : "hair_1", 
        "purple-hair" : "hair_2", 
        "multi-color" : "hair_3", 
        "red-white" : "hair_4", 
        "cap" : "hair_5" , 
        "helmet" : "hair_6"}

acc_files = {"earring" : "acc_1", 
        "cigar" : "acc_2", 
        "yellow-cigar" : "acc_3", 
        "mask": "acc_4", 
        "yellow-dot" : "acc_5", 
        "smoking" : "acc_6", 
        "headphones" : "acc_7"
       }
```

```python
## Generate Traits

TOTAL_IMAGES = 20 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    new_image = {}

    # For each trait category, select a random trait based on the weightings 
    new_image ["Face"] = random.choices(face, face_weights)[0]
    new_image ["Eyes"] = random.choices(eyes, eyes_weights)[0]
    new_image ["Hair"] = random.choices(hair, hair_weights)[0]
    new_image ["Acc"] = random.choices(acc, acc_weights)[0]

    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
   
print(all_images)
```

```python
# Get Trait Counts

face_count = {}
for item in face:
    face_count[item] = 0
    
eyes_count = {}
for item in eyes:
    eyes_count[item] = 0
    
hair_count = {}
for item in hair:
    hair_count[item] = 0
    
acc_count = {}
for item in acc:
    acc_count[item] = 0
    

for image in all_images:
    face_count[image["Face"]] += 1
    eyes_count[image["Eyes"]] += 1
    hair_count[image["Hair"]] += 1
    acc_count[image["Acc"]] += 1
    
    
print(face_count)
print(eyes_count)
print(hair_count)
print(acc_count)
```
